```python
import papermill as pm
import os
import sys
```

```python
def clean_uninstall():
    result = os.system("pip uninstall -y ocs-academic-hub ocs-sample-library-preview gql > /dev/null")
    print(f"@clean={'OK' if result == 0 else 'oops'}")
    return result == 0
```

```python
source_dir = ".."
target_dir = "Executed"
!mkdir -p $target_dir
```

```python
main_nbs = [
    "Wind_Farm_Power_Prediction.ipynb",
    "Wind_Farms_Dataset_Quickstart.ipynb",
    "PandasGuide.ipynb",
]


for_ci = {
    "log_output": True,
    "stdout_file": sys.stdout,
    "stderr_file": sys.stderr,
    "progress_bar": False,
    "report_mode": True,
}


def execute_nbs():
    for nb in main_nbs:
        if clean_uninstall():
            print(f">> {nb}...")
            nb_source = f"{source_dir}/{nb}"
            %env NB_SOURCE=$nb_source
            pm.execute_notebook(
                f"{source_dir}/{nb}",
                f"./{target_dir}/{nb}",
                # log_output=True,
                # stdout_file=sys.stdout,
                # stderr_file=sys.stderr,
                # progress_bar=False,
                # report_mode=True,
                **for_ci,
            )
        else:
            print(f"### Could not clean for {nb}, skipping it")


None
```

```python
if True:
    execute_nbs()
```
